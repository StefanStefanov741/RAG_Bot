```python
%xmode verbose
from pdf2image import convert_from_path

# Convert PDF to images
images = convert_from_path('Inputs/A New Jewelflower (Streptanthus, Brassicaceae) From Marin County, California, Usa.pdf')
for i, image in enumerate(images):
    image.save(f'page_{i}.png', 'PNG')

```

```python
import torch
import detectron2
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2 import model_zoo
import layoutparser as lp

# Set up Detectron2 config
cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"))
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5  # set a threshold for predictions
cfg.MODEL.WEIGHTS = "model_final.pth"  # e.g., PubLayNet pre-trained model
predictor = DefaultPredictor(cfg)

# Load an image (converted PDF page)
from PIL import Image
import numpy as np
img = np.array(Image.open('page_0.png'))

# Perform page object detection
outputs = predictor(img)
```
